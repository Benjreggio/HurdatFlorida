```python
# --- 2. Define the name of the database to be created ---
NEW_DB_NAME = "valid_storms"

def create_database_pyodbc(db_host, db_port, db_user, db_password, new_db_name):
    """
    Creates a database using a direct pyodbc connection. This bypasses
    SQLAlchemy's transactional behavior, which is necessary for CREATE DATABASE.
    """
    conn_string = (
        f"DRIVER={{ODBC Driver 18 for SQL Server}};"
        f"SERVER={db_host},{db_port};"
        f"UID={db_user};"
        f"PWD={db_password};"
        f"TrustServerCertificate=yes;"
    )

    connection = None
    try:
        # Connect directly using pyodbc and force autocommit
        connection = pyodbc.connect(conn_string, autocommit=True)
        cursor = connection.cursor()

        print("\nUsing direct pyodbc connection to create database...")
        create_db_sql = f"IF NOT EXISTS (SELECT name FROM sys.databases WHERE name = '{new_db_name}') CREATE DATABASE {new_db_name};"
        
        cursor.execute(create_db_sql)
        print(f"Database '{new_db_name}' created or already exists.")
        
    except Exception as e:
        print("\nAn error occurred during direct database creation:")
        print(e)
    finally:
        if connection:
            connection.close()

def create_user_and_login_pyodbc(db_host, db_port, db_user, db_password, target_db_name, new_user_name, new_user_password):
    """
    Creates a new login and user with read/write permissions using a direct pyodbc connection.
    This also requires an admin user with appropriate permissions.
    """
    conn_string = (
        f"DRIVER={{ODBC Driver 18 for SQL Server}};"
        f"SERVER={db_host},{db_port};"
        f"UID={db_user};"
        f"PWD={db_password};"
        f"TrustServerCertificate=yes;"
    )

    connection = None
    try:
        connection = pyodbc.connect(conn_string, autocommit=True)
        cursor = connection.cursor()

        print(f"\nAttempting to create user and login for '{new_user_name}'...")
        
        # 1. Create a server-level LOGIN
        create_login_sql = f"IF NOT EXISTS (SELECT * FROM sys.sql_logins WHERE name = '{new_user_name}') CREATE LOGIN [{new_user_name}] WITH PASSWORD = '{new_user_password}';"
        cursor.execute(create_login_sql)
        print("Login created or already exists.")
        
        # 2. Switch to the target database and create the USER
        # This is a critical step
        cursor.execute(f"USE {target_db_name};")
        
        create_user_sql = f"IF NOT EXISTS (SELECT * FROM sys.database_principals WHERE name = '{new_user_name}') CREATE USER [{new_user_name}] FOR LOGIN [{new_user_name}];"
        cursor.execute(create_user_sql)
        print("User created or already exists.")
        
        # 3. Grant the user permissions
        grant_perms_sql = f"""
            ALTER ROLE db_datareader ADD MEMBER [{new_user_name}];
            ALTER ROLE db_datawriter ADD MEMBER [{new_user_name}];
        """
        cursor.execute(grant_perms_sql)
        print("Permissions granted to the new user.")
        
    except Exception as e:
        print("\nAn error occurred during user creation:")
        print(e)
    finally:
        if connection:
            connection.close()

# --- 3. Run the database creation function first ---
create_database_pyodbc(DB_HOST, DB_PORT, DB_USER, DB_PASSWORD, NEW_DB_NAME)

# --- 4. Now, create a new user for the new database ---
# You can define the new user's credentials here.
NEW_USER = "new_user"
NEW_PASSWORD = "my_strong_password"
create_user_and_login_pyodbc(DB_HOST, DB_PORT, DB_USER, DB_PASSWORD, NEW_DB_NAME, NEW_USER, NEW_PASSWORD)

# --- 5. Now, connect to the newly created database using SQLAlchemy ---
try:
    # Construct the final connection string with the new database name
    connection_string = (
        f"mssql+pyodbc://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{NEW_DB_NAME}?"
        f"driver=ODBC+Driver+18+for+SQL+Server&TrustServerCertificate=yes"
    )
    
    # Create a new engine for the new database
    engine = create_engine(connection_string, echo=True)
    
    # Verify the connection to the new database
    with engine.connect() as new_connection:
        print(f"\nSuccessfully connected to the new '{NEW_DB_NAME}' database!")
        result = new_connection.execute(text("SELECT DB_NAME() AS CurrentDB;"))
        for row in result:
            print(f"Current database is: {row[0]}")
    
except Exception as e:
    print("\nAn error occurred while connecting to the new database:")
    print(e)
```

```python
def execute_admin_command(db_host, db_port, db_user, db_password, sql_command, db_name="master"):
    """
    Executes a single, non-transactional SQL administrative command
    (e.g., CREATE DATABASE, CREATE LOGIN) using a direct pyodbc connection.

    Args:
        db_host (str): The database server's hostname.
        db_port (int): The database server's port.
        db_user (str): The username for the database connection.
        db_password (str): The password for the database connection.
        sql_command (str): The SQL command to execute.
        db_name (str): The database to connect to. Defaults to 'master'.
    """
    conn_string = (
        f"DRIVER={{ODBC Driver 18 for SQL Server}};"
        f"SERVER={db_host},{db_port};"
        f"DATABASE={db_name};"
        f"UID={db_user};"
        f"PWD={db_password};"
        f"TrustServerCertificate=yes;"
    )

    connection = None
    try:
        # Connect directly using pyodbc and force autocommit
        connection = pyodbc.connect(conn_string, autocommit=True)
        cursor = connection.cursor()

        print(f"Executing command: {sql_command}")
        result = cursor.execute(sql_command)
        for row in result:
            print(row)
        print("Command executed successfully.")
        
    except Exception as e:
        print("\nAn error occurred during command execution:")
        print(e)
    finally:
        if connection:
            connection.close()
            print("Connection closed.")

# --- Example Usage ---

# Define a sample command to create a new table.
# You would use this after your database has been created.
create_user_command = """
    CREATE USER csharpserver FOR LOGIN csharpserver;
    ALTER ROLE db_datareader ADD MEMBER csharpserver;
"""

show_users_command = """
    SELECT name, type_desc FROM sys.database_principals;
"""

# Call the function to run the command.
# We'll use your new database name for this example.
# Make sure your environment variables are set or update the defaults.
execute_admin_command(DB_HOST, DB_PORT, DB_USER, DB_PASSWORD, show_users_command, db_name="VALID_STORMS")
```
